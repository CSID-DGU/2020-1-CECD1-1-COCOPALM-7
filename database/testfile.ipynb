```python
new_keyword_list = list(set(new_keyword)) # 중복 제거한 키워드 리스트
old_keyword_list = list(set(old_keyword))

new_post_sum = [0] * len(new_keyword_list) # 키워드 당 게시글 카운트
old_post_sum = [0] * len(old_keyword_list)

new_agree_sum = [0] * len(new_keyword_list) # 키워드 당 청원 동의수 카운트
old_agree_sum = [0] * len(old_keyword_list) 

new_best_id = [0] * len(new_keyword_list) # 키워드 당 가장 청원 많은 글id
old_best_id = [0] * len(old_keyword_list) 

for index, value in enumerate(new_keyword_list): # 게시글 카운트
    for keyword in new_keyword:
        if(keyword == value):
            new_post_sum[index] += 1
            
for index, value in enumerate(old_keyword_list):
    for keyword in old_keyword:
        if(keyword == value):
            old_post_sum[index] += 1

for list_index, list_value in enumerate(new_keyword_list): # 동의수 카운트
    for index, value in enumerate(new_keyword):
        if(list_value == value):
            new_agree_sum[list_index] += int(new_keyword_agree_count[index])
            
for list_index, list_value in enumerate(old_keyword_list):
    for index, value in enumerate(old_keyword):
        if(list_value == value):
            old_agree_sum[list_index] += int(old_keyword_agree_count[index])

# 변수 막 지음 주의
new_keyword_index_in_keyword = [] # 키워드가 있는 인덱스 저장
for list_index, list_value in enumerate(new_keyword_list): # best_id 찾기
    new_individual_keyword_agree_count = [] # 키워드에 해당하는 동의수 저장
    new_keyword_index_in_keyword = [index for index, value in enumerate(new_keyword) if value == list_value] # 인덱스 다 찾기
#     print(len(new_keyword))
    for findex in new_keyword_index_in_keyword:
        new_individual_keyword_agree_count += [new_keyword_agree_count[findex]] # 개별 키워드에 해당하는 동의 수 저장
    temporary_index = new_individual_keyword_agree_count.index(max(new_individual_keyword_agree_count)) # 동의 수 가장 많은 임시 배열의 인덱스
    real_index = new_keyword_index_in_keyword[temporary_index] # 실제 그 키워드가 있는 인덱스(id에 접근하기 위해)
    new_best_id[list_index] = new_post_id_for_best_keyword[real_index] # id 값을 real_index로 접근하여 키워드만 있는 배열에 저장


old_keyword_index_in_keyword = []  
for list_index, list_value in enumerate(old_keyword_list): 
    old_individual_keyword_agree_count = []
    old_keyword_index_in_keyword = [index for index, value in enumerate(old_keyword) if value == list_value]
    for findex in old_keyword_index_in_keyword:
        old_individual_keyword_agree_count += [old_keyword_agree_count[findex]] 
    temporary_index = old_individual_keyword_agree_count.index(max(old_individual_keyword_agree_count)) 
    real_index = old_keyword_index_in_keyword[temporary_index]
    old_best_id[list_index] = old_post_id_for_best_keyword[real_index]

print(new_best_id)
print(old_best_id)
```
